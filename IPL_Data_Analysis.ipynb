## Number of Macthes won in ipl 2022 by team wise

Matches won by each team

```python
figure = px.bar(data, x = data['match_winner'], title = 'Number of Matches won in 2022')
figure.show()
```

Winning Matching by Cahasing or Defending

```python
data['won_by'] = data['won_by'].map({'Wickets':'Chasing', 'Runs':'Defending'})
won_by = data['won_by'].value_counts()
label = won_by.index
counts = won_by.values
colors = ['red', 'lightgreen']

# Create the pie chart with corrected syntax
fig = go.Figure(data=[go.Pie(labels=label, values=counts)])

fig.update_layout(title_text = 'Number of matches won by defending chasing')
fig.update_traces(hoverinfo='label+percent', textinfo='value', textfont_size=20,
                  marker = dict(colors = colors, line = dict(color = 'black', width = 3)))

```

Comparing bowling of the bowlers in the year 2022

```python
figure = px.bar(data, x = data['best_bowling'], title = 'Best Bowler in IPL 2022')
figure.show()

```
